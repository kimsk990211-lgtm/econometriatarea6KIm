# Tarea: Cambio de Unidades
## Problema 1

Sea el modelo de regresión lineal simple:

$$
y_i = \hat{\beta}_0 + \hat{\beta}_1 x_i + u_i
$$

donde $\hat{\beta}_0$ es el intercepto y $\hat{\beta}_1$ la pendiente estimados por MCO.

Definimos nuevas variables transformadas:

$$
y_i^* = c_1 y_i, \quad x_i^* = c_2 x_i
$$

con $c_2 \neq 0$.

Queremos encontrar los estimadores $\hat{\beta}_0^*$ y $\hat{\beta}_1^*$ en la regresión de $y_i^*$ sobre $x_i^*$.

---



Sabemos que:

$$
\hat{\beta}_1 = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2}
$$

Para las variables transformadas:

$$
\hat{\beta}_1^* = \frac{\sum (x_i^* - \bar{x}^*)(y_i^* - \bar{y}^*)}{\sum (x_i^* - \bar{x}^*)^2}
$$

---



Sabemos que:

$$
x_i^* = c_2 x_i, \quad \bar{x}^* = c_2 \bar{x}
$$

$$
y_i^* = c_1 y_i, \quad \bar{y}^* = c_1 \bar{y}
$$

Entonces:

$$
x_i^* - \bar{x}^* = c_2 (x_i - \bar{x})
$$

$$
y_i^* - \bar{y}^* = c_1 (y_i - \bar{y})
$$

---



Numerador:

$$
\sum (x_i^* - \bar{x}^*)(y_i^* - \bar{y}^*) 
= \sum [c_2(x_i - \bar{x})][c_1(y_i - \bar{y})]
= c_1 c_2 \sum (x_i - \bar{x})(y_i - \bar{y})
$$

Denominador:

$$
\sum (x_i^* - \bar{x}^*)^2 
= \sum [c_2(x_i - \bar{x})]^2 
= c_2^2 \sum (x_i - \bar{x})^2
$$

---



$$
\hat{\beta}_1^* 
= \frac{c_1 c_2 \sum (x_i - \bar{x})(y_i - \bar{y})}{c_2^2 \sum (x_i - \bar{x})^2}
= \frac{c_1}{c_2} \hat{\beta}_1
$$

---



Sabemos que:

$$
\hat{\beta}_0 = \bar{y} - \hat{\beta}_1 \bar{x}
$$

Para las variables transformadas:

$$
\hat{\beta}_0^* = \bar{y}^* - \hat{\beta}_1^* \bar{x}^*
$$

Sustituimos:

$$
\bar{y}^* = c_1 \bar{y}, \quad \bar{x}^* = c_2 \bar{x}
$$

$$
\hat{\beta}_0^* = c_1 \bar{y} - \left(\frac{c_1}{c_2}\hat{\beta}_1\right)(c_2 \bar{x})
$$

$$
= c_1 \bar{y} - c_1 \hat{\beta}_1 \bar{x}
$$

$$
= c_1 (\bar{y} - \hat{\beta}_1 \bar{x})
= c_1 \hat{\beta}_0
$$

---

## ✅ Resultado final

$$
\boxed{
\hat{\beta}_1^* = \frac{c_1}{c_2}\hat{\beta}_1
}
$$

$$
\boxed{
\hat{\beta}_0^* = c_1 \hat{\beta}_0
}
$$

## Problema 2

Sea el modelo original:

$$
y_i = \hat{\beta}_0 + \hat{\beta}_1 x_i + u_i
$$

Definimos una transformación en ambas variables:

$$
y_i^* = y_i + c_1, \quad x_i^* = x_i + c_2
$$

Queremos encontrar los estimadores $\hat{\beta}_0^*$ y $\hat{\beta}_1^*$ en la regresión de $y_i^*$ sobre $x_i^*$.

---

## ✏️ Demostración

Partimos del modelo original:

$$
y_i = \hat{\beta}_0 + \hat{\beta}_1 x_i + u_i
$$

Sumando $c_1$ en ambos lados:

$$
y_i + c_1 = \hat{\beta}_0 + \hat{\beta}_1 x_i + c_1 + u_i
$$

Sustituyendo $y_i^* = y_i + c_1$:

$$
y_i^* = \hat{\beta}_0 + \hat{\beta}_1 x_i + c_1 + u_i
$$

Ahora expresamos $x_i$ en función de $x_i^*$:

$$
x_i = x_i^* - c_2
$$

Sustituyendo en la ecuación:

$$
y_i^* = \hat{\beta}_0 + \hat{\beta}_1 (x_i^* - c_2) + c_1 + u_i
$$

Distribuyendo:

$$
y_i^* = \hat{\beta}_0 + \hat{\beta}_1 x_i^* - \hat{\beta}_1 c_2 + c_1 + u_i
$$

Reordenando términos:

$$
y_i^* = (\hat{\beta}_0 + c_1 - c_2 \hat{\beta}_1) + \hat{\beta}_1 x_i^* + u_i
$$

---

## ✅ Identificación de coeficientes

Comparando con el modelo:

$$
y_i^* = \hat{\beta}_0^* + \hat{\beta}_1^* x_i^* + u_i
$$

se obtiene:

$$
\boxed{
\hat{\beta}_1^* = \hat{\beta}_1
}
$$

$$
\boxed{
\hat{\beta}_0^* = \hat{\beta}_0 + c_1 - c_2 \hat{\beta}_1
}
$$

## Problema 3

Sea el modelo de regresión:

$$
\log(y_i) = \hat{\beta}_0 + \hat{\beta}_1 x_i + u_i
$$

donde $y_i > 0$ para toda $i$.

Definimos una transformación:

$$
y_i^* = \log(c_1 y_i), \quad \text{con } c_1 > 0
$$

Queremos encontrar los estimadores $\hat{\beta}_0^*$ y $\hat{\beta}_1^*$ en la regresión de $y_i^*$ sobre $x_i$.

---

## ✏️ Demostración

Partimos de la propiedad del logaritmo:

$$
\log(c_1 y_i) = \log(c_1) + \log(y_i)
$$

Entonces:

$$
y_i^* = \log(c_1) + \log(y_i)
$$

Sustituyendo el modelo original:

$$
y_i^* = \log(c_1) + \hat{\beta}_0 + \hat{\beta}_1 x_i + u_i
$$

Reordenando términos:

$$
y_i^* = (\hat{\beta}_0 + \log(c_1)) + \hat{\beta}_1 x_i + u_i
$$

---

## ✅ Identificación de coeficientes

Comparando con el modelo:

$$
y_i^* = \hat{\beta}_0^* + \hat{\beta}_1^* x_i + u_i
$$

se obtiene:

$$
\boxed{
\hat{\beta}_1^* = \hat{\beta}_1
}
$$

$$
\boxed{
\hat{\beta}_0^* = \hat{\beta}_0 + \log(c_1)
}
$$

## Problema 4

Supongamos que $x_i > 0$ para toda $i$.

Sea el modelo original:

$$
y_i = \hat{\beta}_0 + \hat{\beta}_1 \log(x_i) + u_i
$$

Ahora definimos una transformación en la variable explicativa:

$$
x_i^* = \log(c_1 x_i), \quad \text{con } c_1 > 0
$$

Queremos encontrar la relación entre $\hat{\beta}_0^*$ y $\hat{\beta}_1^*$ en la regresión de $y_i$ sobre $\log(c_1 x_i)$.

---

## ✏️ Demostración

Partimos de la propiedad del logaritmo:

$$
\log(c_1 x_i) = \log(c_1) + \log(x_i)
$$

Por lo tanto:

$$
x_i^* = \log(c_1) + \log(x_i)
$$

Despejamos $\log(x_i)$:

$$
\log(x_i) = x_i^* - \log(c_1)
$$

Sustituyendo en el modelo original:

$$
y_i = \hat{\beta}_0 + \hat{\beta}_1 (x_i^* - \log(c_1)) + u_i
$$

Distribuyendo:

$$
y_i = \hat{\beta}_0 + \hat{\beta}_1 x_i^* - \hat{\beta}_1 \log(c_1) + u_i
$$

Reordenando:

$$
y_i = \left(\hat{\beta}_0 - \hat{\beta}_1 \log(c_1)\right) + \hat{\beta}_1 x_i^* + u_i
$$

---

## ✅ Identificación de coeficientes

Comparando con el modelo:

$$
y_i = \hat{\beta}_0^* + \hat{\beta}_1^* x_i^* + u_i
$$

se obtiene:

$$
\boxed{
\hat{\beta}_1^* = \hat{\beta}_1
}
$$

$$
\boxed{
\hat{\beta}_0^* = \hat{\beta}_0 - \hat{\beta}_1 \log(c_1)
}
$$
